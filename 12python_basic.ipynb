### 라인 플롯(line plot) - title(), x 축의 위치(tick), show()

- x축의 위치를 x틱이라고 합니다. default는 0,1....N-1범위의 index array입니다.

```python
plt.title("Plot")
plt.plot([1,4,9,16])
plt.show()
```

```python
global_font = {'family':'Batang', 'size':12, 'weight': 'bold'}

# 폰트 설정
mpl.rc('font',**global_font) # pass in the font divt as kwargs
# mpl.rc('font', family' = 'Batang', 'size':=24, 'weight': 'bold')와 같이 동작합니다.
```

#### 기본 폰트 확인

```python
plt.title('한글 제목')
plt.plot([10,20,30,40], [1,4,9,16])
plt.xlabel("엑스축 레이블")
plt.ylabel("와이축축 레이블")
plt.show()
```

#### 개별 폰트 설정하기

```python
title_font = {'family' : 'Batang',
              'size' : 24,
              'color' : 'black'}
xlabel_font = {'family' : 'Batang',
              'size' : 18,
              'weight' : 'bold',
              'color' : 'darkred'}
ylabel_font = {'family' : 'Batang',
              'size' : 12,
              'weight' : 'light',
              'color' : 'blue'}

plt.plot([10,20,30,40], [1,4,9,16])
plt.title('한글 제목', fontdict=title_font)
plt.xlabel('엑스 축', fontdict=xlabel_font)
plt.ylabel('와이 축', fontdict=ylabel_font)
plt.show()
```

### 틱 설정하기

```python
import numpy as np
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label 설정")
plt.plot(X,C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1,0,+1])
plt.show()
```

#### 

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("Tex, 문자열로 tic label 정의")
plt.plot(X,C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,+1], ["Low", "Zero", "High"])
plt.show()
```

### 여러 개의 선 그리기

```python
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t,t,'r--',t,0.5*t**2,'bs:',t,0.2*t**3,'g^-')
plt.show()
```

```python
plt.title("복수의 plot 명령을 한 그림에서 표현")
style1 =  {"c":"b", "lw":5, "ls": "--", "marker":"o","ms":15,"mec":"g","mew":5,"mfc":"r"}
style2 =  {"c":"k", "lw":3, "ls": ":", "marker":"s","ms":10,"mec":"m","mew":5,"mfc":"c"}

plt.plot([1,4,9,16], **style1)
plt.plot([9,16,4,1], **style2)

plt.show()
```

### 범례(Legend)
여러개의 라인 플롯을 동시에 그리는 경우에는 각 선이 무슨 자료를 표시하는지를  
보여주기 위해 legend 명령으로 범례를 추가할 수 있습니다. 범례의 위치는 자동으로 정해지지만  
수동으로 설정하고 싶으면 loc키워드 인수를 사용합니다.

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S= np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2) # upper left
plt.show()
```

### x,y 축 label

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S= np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Cosine Plot")
plt.show()
```

### Figure

#### Figure, Axes 객체를 만드는 방법은 pyplot.subplot()을 사용하는 것입니다.
그후 Axes.plot()를 활용하여 Axes에 데이터를 그립니다.

```python
fig, ax = plt.subplots() # Create a figure contanining a single axes.
ax.plot([1,2,3,4],[1,4,2,3]); # Plot some data on the axes
```

#### Figure는 전체 그림을 의미하는 객체 입니다.

```python
fig = plt.figure() # an empty figure with no Axes
fig, ax = plt.subplots() # a figure with a single Axes
fig, axs = plt.subplots(2,2) # a figure with a 2x2 grid of Axes
```

#### plt.plot()

```python
np.random.seed(0)
f1 = plt.figure(figsize=(10,2))
plt.title("figure size : (10,2)")
plt.plot(np.random.randn(100))
plt.show()
```

#### Figure 객체 얻기 plt.gcf()

```python
f1 = plt.figure(1)
plt.title("현재의 Figure 객체")
plt.plot([1,2,3,4], 'ro:')
f2 = plt.gcf()
print(f1, id(f1))
print(f2, id(f2))
plt.show()
```

#### 다음과 같은 그래프를 그려보세요.
- title: 섭씨온도와 화씨온도
- xlabel: 화씨 온도
- ylabel: 섭씨 온도
- x 값은 화씨로 32이상 ~ 100미만의 정수 y값은 섭씨로 ℃ = (℉-32)* 5/9 를 통해 구합니다.

```python
F = np.arange(32,100)
C = (F-32)*5/9
plt.plot(F,C,'r-.')
plt.title("섭씨온도와 화씨온도")
plt.xlabel("화씨 온도")
plt.ylabel("섭씨 온도")
plt.show()
```
