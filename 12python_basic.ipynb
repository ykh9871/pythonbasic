그때는 다음처럼 %matplotlib 매직 명령으로 주피터 내부에 그림을 표시하도록 지정해야합니다.

### 라인 플롯(line plot) - title(), x 축의 위치(tick), show()
- title()는 플롯의 제목입니다.
- x축의 위치를 x틱이라고 합니다. default는 0,1....N-1범위의 index array입니다.
- x tick을 지정하려면 두 개의 같은 길이의 리스트ㅡ 혹은 배열 자려를 넣으면 됩니다.
- show는 플롯을 그림으로 표시합니다.

```python
plt.title("Plot")
plt.plot([1,4,9,16])
plt.show()
```

```python
plt.title("x ticks")
plt.plot([10,20,30,40], [1,4,9,16])
plt.show()
```

### 한글 폰트 사용하기
- 한글 제목을 사용하기 위해 title()의 인수로 한글 문자열을 전달하면 아래처럼 제대로 표현되지 않음을 확인할 수 있습니다.

```python
plt.title("한글 제목")
plt.plot([10,20,30,40], [1,4,9,16])
plt.show()
```

#### Font 설치하기
1. 폰트를 다운로드 받습니다.
    - 무료 폰트 사이트 : [https://noonnu.cc/]
2. 다운로드 받은 폰트를 실행하면 설치할 수 있습니다.

#### font_manage
- 설치된 폰트를 조회할 수 있습니다.

```python
import matplotlib.font_manager
font_list = matplotlib.font_manager.findSystemFonts(fontpaths=None, fontext='ttf')
[matplotlib.font_manager.FontProperties(fname=font).get_name() for font in font_list if 'NanumGothic' in font]
```

#### Font scope
설치된 폰트를 사용하는 방벅은 크게 두가지 입니다.
- mpl.rc('font', family='폰트명')메서드를 통한 설정으로 전체 범위에 대해 적용하기  
위 코드는 mpl.rcParams["font.family"] = '폰트명' 과 같습니다.
- fontdict keyword argument를 사용하여 세부적으로 적용하는 방벅

```python
mpl.rcParams["font.family"] = 'NanumGothic'
# mpl.rc('font',family='NanumGothic')
```
